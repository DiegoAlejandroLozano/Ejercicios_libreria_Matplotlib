**Ejercicio 4**

Escribir una función que reciba una serie de Pandas con el número de ventas de un producto durante los meses de un trimestre y un título y cree un diagrama de sectores con las ventas en formato png con el titulo dado. El diagrama debe guardarse en un fichero con formato png y el título dado.

**Solución**

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def guardar_imagen(serie, titulo_imagen: str):
    fig, axes = plt.subplots()
    axes.pie(serie.values, labels=serie.index, shadow=True, textprops={'color':'black'}, autopct='%.1f%%')
    plt.title(titulo_imagen, color='black', fontweight='bold', fontsize=20)
    plt.savefig(titulo_imagen + '.png')
    plt.show()
    
```

```python
datos = {
    'Enero':200,
    'Febrero':150,
    'Marzo':175
}
ventas = pd.Series(datos)
```

```python
guardar_imagen(ventas, 'Ventas')
```

**Ejercicio 1**

Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla un diagrama de líneas con la evolución de las ventas.

**Solución**

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def graficar_ventas(ventas, anios):
    fig, axes = plt.subplots(layout='constrained')
    axes.plot(anios, ventas, marker='o')
    axes.grid()
    axes.set_xlabel('Años')
    axes.set_ylabel('Ventas')
    plt.xticks(anios, rotation=45)
    plt.yticks(ventas)
    plt.show()
```

```python
#Ingreso de registros
n = int(input('Cantidad de registros: '))
ventas = []
anios = []
for _ in range(n):
    anio = int(input('Ingrese el año: '))
    venta = float(input('Ingrese la venta: '))
    anios.append(anio)
    ventas.append(venta)
print(ventas)
print(anios)
```

```python
graficar_ventas(ventas, anios)
```
